```python
import numpy as np
import pandas as pd
import scipy as sp
from scipy import stats

from matplotlib import pyplot as plt
import seaborn as sns
sns.set()
%precision 3
%matplotlib inline
```

```python
population = stats.norm(loc=4,scale=0.8)
```

```python
sample_mean_array = np.zeros(10000)
# sample_mean_array
```

```python
np.random.seed(1)
for i in range(0,10000):
    sample = population.rvs(size=100)#標本抽出をする関数
    sample_mean_array[i] = sp.mean(sample)
```

```python
sp.mean(sample_mean_array)
```

```python
sp.std(sample_mean_array, ddof=1)
```

```python
size_array = np.arange(start=10, stop=10100, step=100)
# size_array
```

```python
sample_mean_array_size = np.zeros(len(size_array))
```

```python
np.random.seed(1)
for i in range(0, len(size_array)):
    sample = population.rvs(size = size_array[i])
    sample_mean_array_size[i] = sp.mean(sample)
```

```python
plt.plot(size_array, sample_mean_array_size)
plt.xlabel("sample size")
plt.ylabel("sample mean")
```

```python
def calc_sample_mean(size, n_trial): # サンプルサイズと試行回数
    sample_mean_array = np.zeros(n_trial)
    for i in range(0, n_trial):
        sample = population.rvs(size=size)
        sample_mean_array[i] = sp.mean(sample)
    return(sample_mean_array)
```

```python
calc_sample_mean(10000, 2)
```
